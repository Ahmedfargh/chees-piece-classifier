```python
def load_img_vs_class():
    directorys={}
    for root, folders, files in os.walk(os.getcwd()):
         for filename in files:
            #directory.appen({filename})
            if filename.endswith(".png"):
                directorys[root+"\\"+filename]=root.split("\\")[7]
                #print(root+" Contain"+filename)
    return directorys
    
```

```python
def map_one_Hot(value):
    vector=[0]*6
    i=0
    for item in classes:
        if value==item:
            vector[i]=1
        i+=1
    return np.array(vector)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
from keras.models import Sequential
from keras.layers import Dense,Flatten,Dropout,Conv2D,AveragePooling2D
from keras.utils import *
import keras
import tensorflow as tf
model=Sequential()
model.add(Conv2D(128,kernel_size=(10,10),input_shape=(85,85,1),activation="relu"))
model.add(MaxPooling2D((10,10)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(Flatten())
model.add(Dense(512,activation="relu"))
model.add(Dense(512,activation="relu"))
model.add(Dense(512,activation="relu"))
model.add(Dense(512,activation="relu"))
model.add(Dense(6,activation="softmax"))
model.compile(optimizer='adam',loss="sparse_categorical_crossentropy",metrics=["accuracy"])
model.summary()
```

```python
#x_train = np.asarray(x_train).astype("float32")
items=[]
for item in x_train:
    items.append(tf.convert_to_tensor(item, dtype=tf.float32))
x_train=np.asarray(items)
x_train=tf.convert_to_tensor(x_train, dtype=tf.float32)
items=[]
for item in y_train:
    items.append(tf.convert_to_tensor(item, dtype=tf.float32))
y_train=np.asarray(items)
y_train=tf.convert_to_tensor(y_train, dtype=tf.float32)
print(type(x_train))
print(type(y_train))
#y_train=tf.convert_to_tensor(y_train, dtype=tf.float32)
model.fit(x_train,y_train,epochs=10)
```

```python
items=[]
for item in x_test:
    items.append(tf.convert_to_tensor(item, dtype=tf.float32))
x_test=np.asarray(items)
x_test=tf.convert_to_tensor(x_train, dtype=tf.float32)
items=[]
for item in y_test:
    items.append(tf.convert_to_tensor(item, dtype=tf.float32))
y_test=np.asarray(items)
y_test=tf.convert_to_tensor(y_train, dtype=tf.float32)
model.evaluate(x_test,y_test)
```
